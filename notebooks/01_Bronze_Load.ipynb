```python
from pyspark.sql.functions import to_timestamp

silver_orders = (
    orders_df
    .withColumn(
        "order_purchase_timestamp",
        to_timestamp("order_purchase_timestamp", "M/d/yyyy H:mm")
    )
    .withColumn(
        "order_approved_at",
        to_timestamp("order_approved_at", "M/d/yyyy H:mm")
    )
    .withColumn(
        "order_delivered_carrier_date",
        to_timestamp("order_delivered_carrier_date", "M/d/yyyy H:mm")
    )
    .withColumn(
        "order_delivered_customer_date",
        to_timestamp("order_delivered_customer_date", "M/d/yyyy H:mm")
    )
    .withColumn(
        "order_estimated_delivery_date",
        to_timestamp("order_estimated_delivery_date", "M/d/yyyy H:mm")
    )
)
```

```python
orders_df.select(
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
).show(10, truncate=False)
```

```python
silver_orders.printSchema()
```

**KPIs Orders**

```python
#approval_time
silver_orders = silver_orders.withColumn(
    "approval_days",
    datediff(
        "order_approved_at",
        "order_purchase_timestamp"
    )
)
```

```python
#warehouse processing
silver_orders = silver_orders.withColumn(
    "carrier_pickup_days",
    datediff(
        "order_delivered_carrier_date",
        "order_approved_at"
    )
)
```

```python
#shipping_time
silver_orders = silver_orders.withColumn(
    "shipping_days",
    datediff(
        "order_delivered_customer_date",
        "order_delivered_carrier_date"
    )
)
```

```python
#total delivery time
silver_orders = silver_orders.withColumn(
    "total_delivery_days",
    datediff(
        "order_delivered_customer_date",
        "order_purchase_timestamp"
    )
)
```

```python
#did we delivered on time?
from pyspark.sql.functions import to_date

silver_orders = silver_orders.withColumn(
    "delivery_status",
    when(
        to_date(col("order_delivered_customer_date")) < to_date(col("order_estimated_delivery_date")),
        "Early"
    ).when(
        to_date(col("order_delivered_customer_date")) == to_date(col("order_estimated_delivery_date")),
        "On Time"
    ).otherwise("Late")
)
```

```python
#validate
silver_orders.select(
    "order_purchase_timestamp",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
    "approval_days",
    "shipping_days",
    "total_delivery_days",
    "delivery_status"
).show(10, truncate=False)
```

```python
silver_orders.select(
    "approval_days",
    "carrier_pickup_days",
    "shipping_days",
    "total_delivery_days",
    "delivery_status"
).show(20, truncate=False)
```

```python
silver_orders.filter(
    col("order_delivered_customer_date").isNull()
).select(
    "order_status",
    "order_delivered_customer_date",
    "shipping_days",
    "total_delivery_days"
).show(20, truncate=False)
```

```python
silver_orders.groupBy("delivery_status").count().show()
```

```python
silver_orders.filter(col("total_delivery_days") < 0).count()
```

```python
from pyspark.sql.functions import countDistinct

silver_orders.select(
    count("*").alias("total_rows"),
    countDistinct("order_id").alias("unique_orders")
).show()
```

**Step 2: Apply Transformations**

```python
from pyspark.sql.functions import *
from pyspark.sql.types import DecimalType

silver_order_items = (
    order_items_df
    .withColumn(
        "shipping_limit_date",
        to_timestamp(
            "shipping_limit_date",
            "yyyy-MM-dd HH:mm:ss"
        )
    )
    .withColumn(
        "price",
        col("price").cast(DecimalType(10,2))
    )
    .withColumn(
        "freight_value",
        col("freight_value").cast(DecimalType(10,2))
    )
    .withColumn(
        "total_item_cost",
        col("price") + col("freight_value")
    )
)
```
