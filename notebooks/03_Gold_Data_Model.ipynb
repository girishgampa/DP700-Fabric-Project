```python
from pyspark.sql.functions import col, to_date, initcap

# Load tables
silver_orders = spark.table("silver_orders")
silver_order_items = spark.table("silver_order_items")
fact_review_summary = spark.table("fact_review_summary")

# Create Fact Sales
fact_sales = (
    silver_order_items.alias("oi")
    .join(
        silver_orders.alias("o"),
        on="order_id",
        how="inner"
    )
    .join(
        fact_review_summary.alias("r"),
        on="order_id",
        how="left"
    )
    .select(

        # Keys
        col("order_id"),
        col("oi.order_item_id"),
        col("o.customer_id"),
        col("oi.product_id"),
        col("oi.seller_id"),

        # Order Information
        col("o.order_purchase_timestamp"),
        to_date(col("o.order_purchase_timestamp")).alias("order_purchase_date"),
        col("o.order_status"),
        initcap(col("o.order_status")).alias("delivery_status"),

        # Sales Metrics
        col("oi.price"),
        col("oi.freight_value"),
        col("oi.total_item_cost"),

        # Review Metrics
        col("r.avg_review_score"),
        col("r.review_count"),

        # Delivery Dates
        col("o.order_approved_at"),
        col("o.order_delivered_carrier_date"),
        col("o.order_delivered_customer_date"),
        col("o.order_estimated_delivery_date"),

        # Shipping
        col("oi.shipping_limit_date")
    )
)
```

```python
# Save Gold Fact Table
fact_sales.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("fact_sales")

print("✅ fact_sales created successfully!")
```

**Operations metrics, creating delivery days and performance columns**

```python
from pyspark.sql.functions import (
    col,
    to_date,
    initcap,
    datediff,
    when,
    lit
)

# Load tables
silver_orders = spark.table("silver_orders")
silver_order_items = spark.table("silver_order_items")
fact_review_summary = spark.table("fact_review_summary")

# Create Fact Sales
fact_sales = (
    silver_order_items.alias("oi")
    .join(
        silver_orders.alias("o"),
        on="order_id",
        how="inner"
    )
    .join(
        fact_review_summary.alias("r"),
        on="order_id",
        how="left"
    )
    .select(

        # ==========================
        # Keys
        # ==========================
        col("order_id"),
        col("oi.order_item_id"),
        col("o.customer_id"),
        col("oi.product_id"),
        col("oi.seller_id"),

        # ==========================
        # Order Information
        # ==========================
        col("o.order_purchase_timestamp"),
        to_date(col("o.order_purchase_timestamp")).alias("order_purchase_date"),
        col("o.order_status"),
        initcap(col("o.order_status")).alias("delivery_status"),

        # ==========================
        # Sales Metrics
        # ==========================
        col("oi.price"),
        col("oi.freight_value"),
        col("oi.total_item_cost"),

        # ==========================
        # Review Metrics
        # ==========================
        col("r.avg_review_score"),
        col("r.review_count"),

        # ==========================
        # Delivery Dates
        # ==========================
        col("o.order_approved_at"),
        col("o.order_delivered_carrier_date"),
        col("o.order_delivered_customer_date"),
        col("o.order_estimated_delivery_date"),

        # ==========================
        # Shipping Information
        # ==========================
        col("oi.shipping_limit_date")
    )

    # ==========================
    # Derived Columns
    # ==========================

    .withColumn(
        "delivery_days",
        datediff(
            col("order_delivered_customer_date"),
            col("order_purchase_date")
        )
    )

    .withColumn(
        "delivery_performance",
        when(
            col("order_delivered_customer_date").isNull(),
            lit("Not Delivered")
        )
        .when(
            to_date(col("order_delivered_customer_date")) < to_date(col("order_estimated_delivery_date")),
            lit("Early")
        )
        .when(
            to_date(col("order_delivered_customer_date")) == to_date(col("order_estimated_delivery_date")),
            lit("On Time")
        )
        .otherwise(
            lit("Late")
        )
    )
)
```

```python
# Save Fact Sales
(
    fact_sales.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("fact_sales")
)

print("✅ fact_sales created successfully!")
```
